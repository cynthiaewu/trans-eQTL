```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def get_mean_errors(type1_arrays):
    means = [np.mean(i) for i in type1_arrays]
    errors = []
    for i in range(len(means)):
        errors.append([means[i] - np.percentile(type1_arrays[i], 25), np.percentile(type1_arrays[i], 75) - means[i]])
    errors = np.array(errors).T
    return means, errors
```
