```python
import scipy.stats as ss
from scipy.stats import gamma
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 5000)
a=2 
plt.figure(figsize=(12, 8), dpi=150)

plt.plot(x, gamma.pdf(x, a), lw=5, alpha=0.6, label='target genes')
plt.plot(x, expon.pdf(x), 'r-', lw=5, alpha=0.6, label='non-target genes')
plt.legend()
plt.xlabel('-log(pvalues)')
plt.ylabel('')
```
