## Distribution of Betas

```python
all_betas = np.loadtxt('/storage/cynthiawu/trans_eQTL/Scripts/Test_betas/all_betas.txt')
```

```python
iterations = 100
x = []
start = -1
for i in range(20):
    x += [start] * iterations
    start += 0.1
    start = float("{:.2f}".format(start))
    if start == 0:
        start = 0.1

all_betas_trans = all_betas.T
y = all_betas_trans.flatten()
plt.scatter(x, y)
```
