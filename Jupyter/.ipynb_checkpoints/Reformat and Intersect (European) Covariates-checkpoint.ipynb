```python
import pandas as pd
```

```python
covariates = pd.read_csv('/storage/polo/GTEx_v8_covariates/gtex_v8_covariates.txt', sep='\t', index_col=0)
```

```python
covariatesT = covariates.transpose()
```

```python
pca = pd.read_csv('/storage/cynthiawu/trans_eQTL/GTex.v8/GTEx_1KG_merged.1kg.pca.evec', index_col=0, sep=' ')
```

```python
pcaT = pca.transpose()
```

```python
pcaT_col = set(pcaT.columns)
covariatesT_col = set(covariatesT.columns)
intersect = pcaT_col.intersection(covariatesT_col)
```

```python
intersect_cov = covariatesT[intersect].append(pcaT[intersect])
```

```python
intersect_cov = intersect_cov.rename(index={'Unnamed: 11': "Race"})
```

```python
european_cov = (intersect_cov.transpose().loc[intersect_cov.transpose()['Race'] == 'European']).transpose()
```
