```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def calculate_type1(samplesize, fdr_sig_threshold):
    pval = []
    for i in range(100):
        fname = f'/gymreklab-tscc/cynthiawu/Test_nullsnps/simulate_eqtls_only/FastMultivariate/Single_eqtl/SampleSize{samplesize}/SingleParameter/numTarget_10/Beta_0/Simulation_{i}/CPMAx/gene-snp-eqtl_cpmax_pvalues_1.0'
        with open(fname, 'r') as f:
            f.readline()
            line = f.readline()
            pval.append(float(line.strip().split('\t')[1]))
    return sum(i < fdr_sig_threshold for i in pval)/len(pval)
    
```

```python
sum(i < 0.05 for i in pval)
```

```python
sig_threshold = 0.05
num_snps_real = 10000
fdr_sig_threshold = sig_threshold/num_snps_real
```

```python
calculate_type1(500, fdr_sig_threshold)
```

```python
# plt.figure(figsize=(10,7))
plt.figure(figsize=(12, 8), dpi=150)



# beta_values = [ 0.01, 0.05, 0.1]
samplesize = [100, 500, 1000, 5000, 10000]
target_colors = {
#     1:"green",
   10: "orange",
    100: "red",
    1000: "blue"
}

sig_threshold = 0.05
num_snps_real = 10000
fdr_sig_threshold = sig_threshold/num_snps_real

errors = []
for s in samplesize:
    errors.append(calculate_type1(s, fdr_sig_threshold))

plt.plot(samplesize, errors, marker='o', linestyle='solid')

      
plt.title('Sample size vs Type 1 error for CPMA')
plt.xlabel('Samplesize')
plt.ylabel('Type 1 error')
plt.ylim(-0.1, 1.1)
# plt.xscale("log")
# plt.legend()

# plt.savefig('/storage/cynthiawu/trans_eQTL/Scripts/Figures/supplementaryfig_1b.pdf') 
```
