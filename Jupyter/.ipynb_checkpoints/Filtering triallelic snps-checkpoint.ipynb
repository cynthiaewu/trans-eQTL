```python
triallelic = set()
with open('/storage/cynthiawu/trans_eQTL/GTex.v8/Run4_080221/Nerve-Tibial/triallelic_snps', 'r') as f:
    for line in f:
        triallelic.add(line.rstrip())
```

```python
for i in range(12, 23):

    notri_f = open(f'/storage/cynthiawu/trans_eQTL/GTex.v8/Run4_080221/Nerve-Tibial/chr{i}/CPMA/cpma_chr{i}_notriallelic', 'w')
    with open(f'/storage/cynthiawu/trans_eQTL/GTex.v8/Run4_080221/Nerve-Tibial/chr{i}/CPMA/cpma_chr{i}', 'r') as f:
        for line in f:
            snp = (line.rstrip().split())[0]
            if snp not in triallelic:
                notri_f.write(line)
```
