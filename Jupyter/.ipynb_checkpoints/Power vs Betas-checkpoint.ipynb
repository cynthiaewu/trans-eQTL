## Distribution of Betas

```python
all_betas = np.loadtxt('/storage/cynthiawu/trans_eQTL/Scripts/Test_betas/all_betas.txt')
```

```python
iterations = 100
x = []
start = -1
for i in range(20):
    x += [start] * iterations
    start += 0.1
    start = float("{:.2f}".format(start))
    if start == 0:
        start = 0.1

all_betas_trans = all_betas.T
y = all_betas_trans.flatten()
plt.scatter(x, y)
```

```python
#plt.figure(figsize=(15,11))
fig, ax = plt.subplots(figsize=(15,10))
ax.boxplot(all_betas)
plt.xticks(range(1,21), betas)
plt.xlabel('Actual Beta Value')
plt.ylabel('Predicted Beta Value')
plt.title('Predicted betas vs Actual simulated betas')
plt.show()
```

# Sample Size 100

```python
samplesize100_lowtargets = pd.read_csv('/storage/cynthiawu/trans_eQTL/Scripts/Test_nullsnps/simulate_eqtls_only/LowTargets/SampleSize100/Simulation_0/CPMA/gene-snp-eqtl_power', sep='\t')
samplesize100_medtargets = pd.read_csv('/storage/cynthiawu/trans_eQTL/Scripts/Test_nullsnps/simulate_eqtls_only/mediumTarget/SampleSize100/Simulation_0/CPMA/gene-snp-eqtl_power', sep='\t')
samplesize100_medhightargets = pd.read_csv('/storage/cynthiawu/trans_eQTL/Scripts/Test_nullsnps/simulate_eqtls_only/medhighTarget/SampleSize100/Simulation_0/CPMA/gene-snp-eqtl_power', sep='\t')
samplesize100_hightargets = pd.read_csv('/storage/cynthiawu/trans_eQTL/Scripts/Test_nullsnps/simulate_eqtls_only/HigherTargets/SampleSize100/Simulation_0/CPMA/gene-snp-eqtl_power', sep='\t')
```

```python
num_targets_100 = list(samplesize100_lowtargets['#target_genes']) + list(samplesize100_medtargets['#target_genes']) + list(samplesize100_hightargets['#target_genes'])
beta_values_100 = list(samplesize100_lowtargets['beta_value']) + list(samplesize100_medtargets['beta_value']) + list(samplesize100_hightargets['beta_value'])
power_100 = list(samplesize100_lowtargets['power']) + list(samplesize100_medtargets['power']) + list(samplesize100_hightargets['power'])
zipped_values = sorted(zip(num_targets_100, power_100, beta_values_100), key=lambda x: x[2])
plt.figure(figsize=(10,7))
current_beta = zipped_values[1][2]
current_targets = []
current_power = []
for value in zipped_values: 
    if value[2] == current_beta:
        current_targets.append(value[0])
        current_power.append(value[1])
    else:
        plt.plot(current_targets, current_power, label=f'beta={current_beta}')
        current_beta = value[2]
        current_targets = [value[0]]
        current_power = [value[1]]
plt.plot(current_targets, current_power, label=f'beta={current_beta}')
plt.xlabel('#Target Genes')
plt.ylabel('Power')
plt.title('Power vs # Target Genes for Sample Size=100, genes=15000, allele freq = 0.5, cov=identity, simulations=100')
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.legend()
```
